```python
import pandas as pd
```

```python
# read in the crime dataset
file = "LocalCrimeOneYr.csv"
df_crime = pd.read_csv(file)
new_df=df_crime
new_df.head()
```

```python
# filter the city name and create as new column
new_df['CityDept'], new_df['Dept'] = df_crime['Agency'].str.rsplit(' Police').str
new_df.head()
```

```python
# select the columns we need filter out the rows that is not a police department
new_df_col = new_df[['CityDept','Population','Agency', 'Violent crime total','Robbery', 'Property crime total','Burglary','Motor vehicle theft']].copy()
new_df_col1=new_df_col[~new_df_col.Agency.str.contains("Sheriff")== True]
new_df_col1.head()
```

```python

# assign new names to the columns selected
renamed_df = new_df_col1.rename(columns={"Violent crime total":"Violent_crime", "Property crime total":"Property_crime","Motor vehicle theft":"MotorVehicle_theft"})
renamed_df.head()
```

```python
# reset the index on this dataframe
new_df_crime = renamed_df.reset_index(drop=True)
new_df_crime.head()
```

```python
# read the county information dataset
file1= "ca_county_FiveYr.csv"
df_countydata = pd.read_csv(file1)
df_countydata.head()
```

```python
# filter the city name and create as new column
new_df1=df_countydata
new_df1['CityDept'], new_df['Dept1'] = df_countydata['agency_name'].str.rsplit(' Police').str
new_df1.head()
```

```python
# filter for year 2014 only
yr2014 = new_df1['year']== 2014
new_df1yr= new_df1[yr2014]
new_df1yr.head()
```

```python
#select the columns we need, and filter data for individual cities

new_df1yr_col = new_df1yr[['CityDept', 'county','agency_name','agency_code', 'bachelors_or_higher','median_income', 'unemployment_rate']].copy()
new_df1yr_colclear = new_df1yr_col[new_df1yr_col.agency_name != "All Combined"]
new_df1yr_colclear.head()
```

```python
# reset the index on this dataframe
new_df_demogrp = new_df1yr_colclear.reset_index(drop=True)
new_df_demogrp.head()
```
